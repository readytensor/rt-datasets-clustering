```python
import numpy as np
import pandas as pd
import os
import json
import pprint
```

```python
dataset_name = 'spam2'
```

```python
input_dir = './raw'
output_dir = './processed'

inp_fname = 'dataset_44_spambase.csv'

outp_fname = os.path.join(output_dir, f'{dataset_name}.csv')
outp_test_key_fname = os.path.join(output_dir, f'{dataset_name}_test_key.csv')
```

# Read Data

```python
data = pd.read_csv(os.path.join(input_dir, inp_fname))
data.head()
```

```python
id_col = "id"
target_col = "class"
```

# Insert Id Column

```python
# insert Id column 
if id_col not in data.columns:
    N = data.shape[0]
    data.insert(0, id_col, np.arange(N))
    print(data.head())

data[id_col] = data[id_col].astype(str)
```

# Shuffle Data

```python
# shuffle data
data = data.sample(frac=1, random_state=42)
data.head()
```

# Save Main Data File

```python
data.drop(target_col,axis=1).to_csv(outp_fname, index=False)
```

# Test Key

```python
data_key = data[[id_col, target_col]].rename(columns={target_col: "__target__"})
data_key.to_csv(outp_test_key_fname, index=False)
```
